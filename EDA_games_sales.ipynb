```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_link = 'https://docs.google.com/spreadsheets/d/1z46qH9il8OfVkYBZMMDUYq_GYHp77_prgkKq6HysZts/edit#gid=1485085913'
data_link_trf = data_link.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(data_link_trf)
df
```

```python
df_cleaned = df.drop_duplicates()
df_cleaned['Release'] = pd.to_datetime(df['Release'])
df_cleaned['Release']
```

```python
for x in ['Name', 'Sales', 'Series', 'Release', 'Genre', 'Developer',
       'Publisher']:
       print(df_cleaned[x].value_counts(), '/n')
```

```python
#download dataset clean
df_cleaned.to_csv('games_sales_cleaned.csv')
```

EDA

```python
#game lama
gameoldtonew = df_cleaned.sort_values(by='Release')
df_gameoldtonew = gameoldtonew[['Name', 'Release']]
df_gameoldtonew.head(10)
```

```python
#game baru
gameoldtonew = df_cleaned[['Name', 'Release']].sort_values('Release', ascending=False)
gameoldtonew.head(10)
```

```python
#publisher terbanyak berdasarkan game
publisher = (df_cleaned.groupby(by=['Publisher'])['Name']
       .count()
       .sort_values(ascending=False)
       .reset_index(name='Jumlah game')
       )
publisher.head(5)
```

```python
developer = (df_cleaned.groupby(by=['Developer'])['Name']
             .nunique()
             .sort_values(ascending=False)
             .reset_index(name='Banyak_game'))
developer.head(5)
```

```python
#series yang memiliki jumlah sales terbanyak
games_sales = (df_cleaned.groupby(by=['Series'])['Sales']
       .sum()
       .sort_values(ascending=False)
       .reset_index(name='Jumlah Sales'))
games_sales.head(5)
```

```python
#series yang merilis game terbanyak
series_game = (df_cleaned.groupby(by=['Series'])['Name']
       .count()
       .sort_values(ascending=False)
       .reset_index(name='banyak game'))
series_game.head(5)
```
